```
import os 
import sys
import json
import numpy as np 
from json import JSONDecodeError  
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import matplotlib.dates as mdates
from scipy import stats
from matplotlib.cbook import boxplot_stats
from scipy.stats import iqr
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
from statsmodels.tsa.stattools import adfuller, kpss
from sklearn import preprocessing
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
from math import sqrt




import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import *
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error
from keras.callbacks import EarlyStopping
from keras.models import load_model
```

```
#uncomment for google drive
#df = pd.read_csv('/content/drive/My Drive/mldm_cw/datasets/billboard_1950-2019_ultimate_mean_2.0.csv',index_col='year')

df = pd.read_csv('billboard_1950-2019_ultimate_mean_2.0.csv',index_col='year')
acoustic = df['acousticness']
print(acoustic.head())
acoustic.plot(title='acousticness')
pyplot.show()
print(len(acoustic))


```

```
TRAIN_SIZE =0.6
#f = open('/content/drive/My Drive/mldm_cw/Model/rmse_results.txt','w')
col = ['acousticness', 'danceability','energy','liveness','loudness', 'valence']
rp=[]

for title in col:
  #transform series to supervised learning
  raw_values= df[title].values
  diff_values = df[title+'_mean_diff']
  diff_values = diff_values.values
  diff_values=diff_values[1:]
  diff_values = diff_values.reshape((-1,1))
  print(diff_values.shape)
  print(type(raw_values))


  # We apply the MinMax scaler from sklearn
  # to normalize data in the (0, 1) interval
  scaler = MinMaxScaler(feature_range = (0, 1))
  diff_values = scaler.fit_transform(diff_values)

  #create train and test datasets
  train_size = int(len(diff_values) * TRAIN_SIZE)
  test_size = len(diff_values) - train_size
  train, test = diff_values[0:train_size], diff_values[train_size:len(diff_values)]

  print(train.shape)
  print(test.shape)

  print("Number of entries (training set, test set): " + str((len(train), len(test))))


  def create_dataset(dataset, window_size = 1):
      data_X, data_Y = [], []
      for i in range(len(dataset) - window_size - 1):
          a = dataset[i:(i + window_size), 0]
          data_X.append(a)
          data_Y.append(dataset[i + window_size, 0])
      return(np.array(data_X), np.array(data_Y))


  def timeseries_to_supervised(data, window_size=1):
    data_X, data_Y = [], []

    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, window_size+1)]
    columns.append(df)
    df = pd.concat(columns, axis=1)
    df.fillna(0, inplace=True)
    df = df.values
    for i in range(len(df)):
      data_X.append(df[i][:-1])
      data_Y.append(df[i][-1])


    data_X,data_Y = np.array(data_X), np.array(data_Y).reshape(-1,1)
    return (data_X, data_Y)

  window_size = 5
  train_X, train_Y = timeseries_to_supervised(train, window_size)
  test_X, test_Y = timeseries_to_supervised(test, window_size)
  print("Original training data shape:")
  print(train_X.shape)
  print(train_Y.shape)


  # Reshape the input data into appropriate form for Keras.
  train_X = np.reshape(train_X, (train_X.shape[0], 1, train_X.shape[1]))
  test_X = np.reshape(test_X, (test_X.shape[0], 1, test_X.shape[1]))
  print("New training data shape:")
  print(train_X.shape)
  
  dropout=0.1
  def fit_model(train_X, train_Y, window_size = 1):
      model = Sequential()
      
      model.add(LSTM(4,input_shape = (1, window_size),dropout=dropout))
      model.add(Dense(1))
      model.compile(loss = "mean_squared_error", 
                    optimizer = "adam")
      model.fit(train_X, 
                train_Y, 
                epochs = 100, 
                batch_size = 1, 
                verbose = 2)
      
      return(model)

  # Fit the first model.
  model1 = fit_model(train_X, train_Y, window_size)
  #model1.save('/content/drive/My Drive/mldm_cw/Model/'+title+'_model.h5')






  def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]


  def predict_and_score(model, X, Y):
      orig_pred=[]
      orig_datas=[]
      # Make predictions on the original scale of the data.
      pred = scaler.inverse_transform(model.predict(X))
      print(pred.shape)
      # Prepare Y data to also be on the original scale for interpretability.
      for i in range(len(Y)):
        yhat = pred[i]
      #UN-DIFFERENCE
        yhat = inverse_difference(raw_values[:42], yhat, len(Y)+1-i)
        orig_pred.append(yhat)
      orig_data = scaler.inverse_transform(Y)
      print(orig_data.shape)
      for i in range(len(Y)):
        y = orig_data[i,:]
        y = inverse_difference(raw_values[:42], y, len(Y)+1-i)
        orig_datas.append(y)
       #Calculate RMSE.
      #f.write(title+' train'+'\n')
      print(title+' train')
      for i in range(len(orig_pred)):
        print(' Predicted=%f, Expected=%f' % (orig_pred[i], orig_datas[i]))
        #f.write('Predicted=%f, Expected=%f\n' % (orig_pred[i], orig_datas[i]))
      score = sqrt(mean_squared_error(orig_datas, orig_pred))
      return(score, orig_pred)

  def predict_and_score_test(model, X, Y):
    orig_pred=[]
    orig_datas=[]
     #Make predictions on the original scale of the data.
    pred = scaler.inverse_transform(model.predict(X))
    print(pred.shape)
     #Prepare Y data to also be on the original scale for interpretability.
    for i in range(len(Y)):
      yhat = pred[i]
    #UN-DIFFERENCE
      yhat = inverse_difference(raw_values, yhat, len(Y)+1-i)
      orig_pred.append(yhat)
    orig_data = scaler.inverse_transform(Y)
    print(orig_data.shape)
    for i in range(len(Y)):
      y = orig_data[i,:]
      y = inverse_difference(raw_values, y, len(Y)+1-i)
      orig_datas.append(y)
    # Calculate RMSE.
    #f.write(title+' test'+'\n')
    print(title+' test')
    for i in range(len(orig_pred)):
      print('Predicted=%f, Expected=%f' % (orig_pred[i], orig_datas[i]))
      #f.write('Predicted=%f, Expected=%f\n' % (orig_pred[i], orig_datas[i]))
    score = sqrt(mean_squared_error(orig_datas, orig_pred))
    return(score, orig_pred)

  rmse_train, train_predict = predict_and_score(model1, train_X, train_Y)
  rmse_test, test_predict = predict_and_score_test(model1, test_X, test_Y)

  print("Training data score: %.4f RMSE" % rmse_train)
  print("Test data score: %.4f RMSE" % rmse_test)
  #f.write("\nTraining data score: %.4f RMSE\n" % rmse_train)
  #f.write("Test data score: %.4f RMSE\n" % rmse_test)
  
  

  # Create the plot.
  plt.figure(figsize = (15, 5))
  train_predict_= [0]+train_predict
  test_predict_ = [0]+([0]*len(train_predict)+test_predict)
  train_predict_ = pd.DataFrame(train_predict_)
  test_predict_ = pd.DataFrame(test_predict_)

  train_predict_.replace(0, np.nan, inplace=True)
  test_predict_.replace(0, np.nan, inplace=True)
  train_predict_.index = pd.RangeIndex(1950,1992)
  test_predict_.index = pd.RangeIndex(1950,2020)
  print(train_predict_.tail())
  plt.plot(df[title], label = 'true value')
  plt.plot(train_predict_,label='training set prediction')
  plt.plot(test_predict_,label='test set prediction')
  plt.legend()
  plt.title(title)
  #plt.savefig('/content/drive/My Drive/mldm_cw/Model/'+title+'.png')
  plt.show()

#f.close()
```
